```python
# TASK 3_1 impute missing values in c with grouped [ by a ] mean of c .
import pandas as pd
import numpy as np
df=pd.DataFrame({'a':np.random.choice([1,2,3,4],50),
 'c':np.random.choice([10,20,np.nan],50)})
#df
df1=df.groupby('a',as_index=False)['c'].transform('mean')
#df1
df['c']=df['c'].fillna(df1['c'])
df
```

```python
#TASK 3_3 Write a function which takes input a name [having first name and last name] and returns only first
#letter of both first and last name as capital

def caps (name):
  name = [i.title() for i in name.split()]
  print(name)

name=input('\nEnter Full name:\n')
name=name.lower()
#print(type(name))
caps(name)
```

```python
#TASK 3_4 A string is encrypted using a dictionary . Write a function which returns the original string , taking input
#the encryted string and the encrypting dictionary 

def conv(str1,dict1):
  #sot = []
  for i in str1:
    for key, value in dict1.items():
      if i==value:
         str1=str1.replace(i,key)
    
  print(str1)
 
str1=input('\nEnter Encrypted input string:\n')
dict1={'a':'@','i':'1','s':'$','n':'#'}
conv(str1,dict1)



```
