```python
#1 Merge the 3 dataset and create 1 view of data.
#You can merge insurance_data.csv and employee_data.csv on
#AGENT_ID You can merge insurance_data.csv and vendor_data.csv on
#VENDOR_ID Note: Use left Outer join as not all claims require Vendor

import pandas as pd
import numpy as np
df1 = pd.read_csv('employee_data.csv')
df2 = pd.read_csv('insurance_data.csv')
df3 = pd.read_csv('vendor_data.csv')
view=pd.merge(df2,df1,on='AGENT_ID',how='left').merge(df3,on='VENDOR_ID',how='left')
view

```

```python
# 3 Business Leader wants to find Top 5 States where we are getting most insurance claims
#for customer belonging to HIGH(H) risk segment?

df2[df2['RISK_SEGMENTATION']=='H'].STATE.value_counts()[:5]
```

```python
# Task #4
#Business wants to create a new variable “COLOCATION” which will have following values IF
#Customer State == Incident State == Agent Address State THEN 1 ELSE 0 Find the mean of this new column

view['COLOCATION']=(view['STATE_x']==view['INCIDENT_STATE']) & (view['INCIDENT_STATE']==view['STATE_y'])
view['COLOCATION']=view.COLOCATION.astype('int')
view.COLOCATION.mean()
```

```python
# Task #5
#Data entry error was detected in the data and you are required to correct it. If for any
#claim transaction “AUTHORITY_CONTACTED” is NOT “Police” and POLICE_AVAILABLE == 1 Then Update “AUTHORITY_CONTACTED” to Police.
Non_police=view[view['AUTHORITY_CONTACTED']!='Police']
update_rows=Non_police[Non_police['POLICE_REPORT_AVAILABLE']==1].index
view.loc[update_rows,'AUTHORITY_CONTACTED']='Police'
view[(view['POLICE_REPORT_AVAILABLE']=='1') & (view['AUTHORITY_CONTACTED']!='Police')]
```

```python
# 6 Task Business wants to check the Claim Amount for deviation for each transaction, they would like you to calculate as follow
#CLAIM_DEVIATION = AVG_CLAIM_AMOUNT_FOR_LAST_30DAYS (same insurance type)
##If the value < 0.5 THEN CLAIM_DEVIATION = 1 ELSE 0
#If there is less than 30 days of transaction history THEN
#-1 Note: LAST_30DAYS does not include current day

view.TXN_DATE_TIME.max()
curr_date=pd.to_datetime('2021-06-30')

view['TXN_DATE_TIME']=pd.to_datetime(view['TXN_DATE_TIME'])

last_30= view.loc[(view['TXN_DATE_TIME'] > '2021-05-30')&(view['TXN_DATE_TIME'] <'2021-06-30')]

avg_claim=last_30.groupby('INSURANCE_TYPE').mean()['CLAIM_AMOUNT'].to_dict()

# defining a function for calculating claim_deviation
def Claim_Deviation(row):
  if(curr_date- row['TXN_DATE_TIME']) < pd.Timedelta(days=30):
    return -1
  Claim_deviation= avg_claim[row['INSURANCE_TYPE']] / row['CLAIM_AMOUNT']
  if Claim_deviation < 0.5:
    return 1
  else:
    return 0

    # Applying the function to the dataframe
view['CLAIM_DEVIATION'] = view.apply(Claim_Deviation,axis=1)

view.CLAIM_DEVIATION.value_counts()
```

```python
# 7 Find All Agents who have worked on more than 2 types of Insurance Claims. Sort them by
#Total Claim Amount Approved under them in descending order

type_c=view.groupby('AGENT_ID')['INSURANCE_TYPE'].nunique()
ins_typ=type_c > 2
multi_type_agent=ins_typ[ins_typ].index
top_agents= view.groupby(['AGENT_ID','AGENT_NAME']).sum().loc[multi_type_agent][['CLAIM_AMOUNT']].sort_values('CLAIM_AMOUNT',ascending=False)
top_agents.join(type_c)
```

```python
# 8 Mobile & Travel Insurance premium are discounted by 10%
#Health and Property Insurance premium are increased by 7%
#Life and Motor Insurance premium are marginally increased by 2%
#What will be overall change in % of the Premium Amount Collected for all these Customer?

old_premium= view['PREMIUM_AMOUNT'].sum()
prem_by_types= view.groupby('INSURANCE_TYPE').sum()['PREMIUM_AMOUNT'].to_dict()
prem_by_types['Mobile']*=0.90
prem_by_types['Travel']*=0.90

prem_by_types['Health']*=1.07
prem_by_types['Property']*=1.07

prem_by_types['Life']*=1.02
prem_by_types['Motor']*=1.02

revise_prem=sum(prem_by_types.values())

(revise_prem-old_premium)*100/old_premium
```

```python
# 9 Business wants to give discount to customer who are loyal and under stress due to Covid
#They have laid down an eligibility Criteria as follow IF CUSTOMER_TENURE > 60 AND EMPLOYMENT_STATUS = “N”
#AND NO_OF_FAMILY_MEMBERS >=4 THEN 1 ELSE 0 Create a new column “ELIGIBLE_FOR_DISCOUNT” and find it mean.


disc_status= (view['TENURE'] > 60)&(view['EMPLOYMENT_STATUS']=='N')&(view['NO_OF_FAMILY_MEMBERS'] >= 4)
view['Eligible_for_discount']= disc_status.astype(int)
view.Eligible_for_discount.mean()
```

```python
# 10 Business wants to check Claim Velocity which is defined as follow CLAIM_VELOCITY = NO_OF_CLAIMS_IN_LAST30DAYS (for the current insurance type)
#/ NO_OF_CLAIMS_IN_LAST3DAYS (for the current insurance type) Note: LAST30DAYS & LAST3DAYS does not include current Day

view.TXN_DATE_TIME.max()
view['CLAIM_STATUS'].value_counts()
curr_date=pd.to_datetime('2021-06-30')

view['TXN_DATE_TIME']=pd.to_datetime(view['TXN_DATE_TIME'])
last_30=view.loc[(view['TXN_DATE_TIME'] > '2021-05-30')&(view['TXN_DATE_TIME']<'2021-06-30')]
last_3=view.loc[(view['TXN_DATE_TIME'] >= '2021-06-27')&(view['TXN_DATE_TIME'] < '2021-06-30')]

# Calculating total approved claims insurance type wise in last 30 days
claims_last_30_days=pd.DataFrame.from_dict(last_30[last_30['CLAIM_STATUS']=='A'].groupby('INSURANCE_TYPE').size().to_dict(),orient='index',columns=['No_of_claims_last_30_days'])
claims_last_3_days=pd.DataFrame.from_dict(last_3[last_30['CLAIM_STATUS']=='A'].groupby('INSURANCE_TYPE').size().to_dict(),orient='index',columns=['No_of_claims_last_3_days'])

total_approved=claims_last_30_days.join(claims_last_3_days)


total_approved['CLAIM_VELOCITY']= total_approved['No_of_claims_last_30_days']/total_approved['No_of_claims_last_3_days']
total_approved
```

```python
# 11 Find all low performing agents i.e. employees who are in the bottom 5 percentile based
#on Claims worked by them.

bottom_5_percentile=view.groupby('AGENT_ID').sum().quantile(0.05)['CLAIM_AMOUNT']

agent_group_data=view.groupby('AGENT_ID').sum()

low_performing= agent_group_data[agent_group_data['CLAIM_AMOUNT']<= bottom_5_percentile][['CLAIM_AMOUNT']]
low_performing
```

```python
#12 Business wants to find all Suspicious Employees (Agents).
#IF TOTAL CLAIM AMOUNT which meet below criteria is >= 15000 THEN AGENT is classified as Suspicious ELSE Not
#CLAIM_STATUS = Approved AND CUSTOMER_RISK_SEGMENTATION = High
#AND INCIDENT_SEVERITY = “Major Loss”
#If Suspicious, then 1 ELSE 0. Find mean of this column.

criteria= (view['CLAIM_STATUS']=='A')&(view['RISK_SEGMENTATION']=='H')&(view['INCIDENT_SEVERITY']=='Major Loss')
criteriamet=view[criteria]
grouped_suspicious=criteriamet.groupby('AGENT_ID').sum()
sus_agents=list(grouped_suspicious[grouped_suspicious['CLAIM_AMOUNT']>15000].index)

df1['SUSPICIOUS']=0
sus_emp= df1['AGENT_ID'].isin(sus_agents)
df1.loc[sus_emp,'SUSPICIOUS']=1
df1['SUSPICIOUS'].mean()
```
